```python
#bus scrapping
def bus_details(href,route_text):
    bus_routes = [] 
    try:
        bus_elements = driver.find_elements(By.CSS_SELECTOR, '.clearfix.bus-item')
        for bus in bus_elements:
            bus_name = WebDriverWait(bus, 30).until(EC.presence_of_element_located((By.CSS_SELECTOR, '.travels.lh-24.f-bold.d-color'))).text
            bus_type = bus.find_element(By.CSS_SELECTOR, '.bus-type').text
            departing_time = bus.find_element(By.CSS_SELECTOR, '.dp-time.f-19.d-color.f-bold').text
            duration = bus.find_element(By.CSS_SELECTOR, '.dur.l-color.lh-24').text
            reaching_time = bus.find_element(By.CSS_SELECTOR, '.bp-time.f-19.d-color.disp-Inline').text
            star_rating_xpath = './/div[@class="rating-sec lh-24"]/div/span'
            star_rating_elements = bus.find_elements(By.XPATH, star_rating_xpath)
            star_rating = star_rating_elements[0].text if star_rating_elements else "N/A"
            price = bus.find_element(By.CSS_SELECTOR, '.seat-fare .f-19').text
            seat_availability = bus.find_element(By.CSS_SELECTOR, '.seat-left').text
            bus_routes.append({'route_link': href,'route_name': route_text,'bus_name': bus_name,'bus_type': bus_type,'departing_time': departing_time,
                    'duration': duration,'reaching_time': reaching_time,'star_rating': star_rating,'price': price,'seat_availability': seat_availability})
        scrolling()
    except:
        print("No bus")
    return bus_routes
#Handiling page
def next_page(page_number):
    try:
        pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
        next_page_button = pagination_container.find_element(By.XPATH, f'.//div[text()="{page_number + 1}"]')
        active_page = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '.DC_117_pageTabs.DC_117_pageActive')))
        # Scroll to the next page button and click it
        actions = ActionChains(driver)
        actions.move_to_element(next_page_button).perform()
        time.sleep(2)
        next_page_button.click()
        print(f'Successfully navigated to page {page_number + 1}')
    except Exception as e:
        print(f'Error navigating to page {page_number + 1}')
```

```python
def scrolling():
    scrolling=True
    while scrolling:
        old_page_source=driver.page_source   
        body=driver.find_element(By.TAG_NAME,"body")
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(10)
        new_page_source=driver.page_source
        if new_page_source==old_page_source:
            scrolling=False
def click_viewbuses():
    try:
    # Wait for all buttons with the specified class name to be present
        buttons = WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.XPATH, "//*[@class='button']")))    
    # Iterate over each button
        for index, button in enumerate(buttons):
            try:
                driver.execute_script("arguments[0].scrollIntoView();", button)
                
                # Wait until the button is clickable
                WebDriverWait(driver, 20).until(EC.element_to_be_clickable(button))
                time.sleep(2)
                # Click the button
                button.click()
                print(f"Button clicked")
            except:
                driver.execute_script("arguments[0].click();", button)
                time.sleep(2)
    except:
        print("No viewbus in the page")
```
